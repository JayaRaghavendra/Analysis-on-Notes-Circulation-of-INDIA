```python
#importing packages

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#loading the dataset 

inc = pd.read_csv("CashCN.csv")

#take a look on your data
inc
```

```python
#let's see the dimensions of the dataset
inc.shape
```

```python
#know the basic info of the like datatypes,datatypes
inc.info()
```

```python
#know your datatypes 
inc.dtypes
```

```python
#let's replace value '0' where data filled with '-'
#so that we can make our analysis effectively and easily

inc = inc.replace('-',0)
```

```python
#check the Null value contained rows

inc[inc.isnull().any(axis=1)]
```

```python
#handle missing by imputing Zero 

inc.fillna(0,inplace=True)
```

```python
inc
```

```python
#let's the check the dtypes which are not numerical 

inc.select_dtypes(include="object").head(2)
```

```python
inc.dtypes
```

```python
#except the year we are changing all columns into float dtype
df = inc.iloc[:,1:].astype(np.float64)
```

```python
#check once to confirm whether we have any null value
df.isna().sum()
```

```python
#adding year columns into new dataframe
df['year'] = inc['year'] 
```

```python
#checkig columns
df.columns
```

```python
#rearranging the columns 
df = df[[ 'year','` 1', '` 2', '` 5', '` 10', '` 20', '` 50', '` 100', '` 200', '` 500','` 1000', '` 2000']]
```

```python
#now wehave cleaned and organised data
df
```

```python
df.info()
```

```python
df.describe()
```

```python
#save this cleaned data for future purposes
df = df.to_csv("Indian banknote demonetisation.csv")
```

```python
#Lets see the corr() of data
plt.figure(figsize=(12,10))
sns.heatmap(df.iloc[:,1:-1].corr(),annot=True,cmap='coolwarm')


#this code is for not getting the top and bottoms of our figure, it's a version issue 

b, t = plt.ylim() # discover the values for bottom and top
b += 0.5 # Add 0.5 to the bottom
t -= 0.5 # Subtract 0.5 from the top
plt.ylim(b, t) # update the ylim(bottom, top) values
plt.show() # ta-da!
```

```python
#lets see the each Note contributons individually by years 

df.plot.bar(figsize=(12,10),subplots=True)
```

```python
#same data in a trend manner 
df.plot(figsize=(8,16),subplots=True)
```

```python
#the best way to understand this kind of data how they are differentiating each other by using area plot 

fig, ax = plt.subplots(figsize=(16,8))
df.plot.area(ax=ax, xticks=df.year.index, rot=90)
ax.set_xticklabels(df.year)

#stackedmbar chart to check all over at once
df.plot.bar(figsize=(16,8),stacked=True)
```

```python
#setting year as index
df=df.set_index('year')
```

```python
df.head(2)
```

```python
#preparing data for 100% area plot
data_perc= df.divide(df.sum(axis=1),axis=0)
```

```python
ax = data_perc.plot.area(figsize=(14,10),stacked=True,rot=90)

plt.title("occupancy of the Notes")
ax.set_ylabel("Percent (%)")

plt.legend(loc=(1.04,0.6))
plt.margins(0,0)
plt.tight_layout()

plt.show()
```
